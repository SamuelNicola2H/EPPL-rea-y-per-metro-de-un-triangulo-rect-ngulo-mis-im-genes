```python
# Calcular Area de triangulo rectangulo
# Variables h, b, A
# Formulas A = b * h / 2
print("Calcular Area de triangulo rectangulo")
h = float(input("Ingrese el valor de la altura: "))
b = float(input("Ingrese el valor de la base: "))
A = b * h / 2
print("El area del triangulo rectangulo es: ", A)
print("Samuel Nicola NL = 40")
```
